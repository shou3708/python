```python
#**********************************************************************
#*
#*  	Name :	彭鍾碩
#*
#*  	Class :	資管四
#*
#*      SID :	S06490001
#*
#*      Program Name : S06490001_hw1.ipynb
#*
#*  	Functions :
#*               (1). 在使用者操作登入時，系統畫面會明確提示要求輸入【身分證號碼】或【手機號碼】其中一個選項。
#*
#*               (2). 在使用者完成輸入後，系統會判讀並驗證客戶輸入的是身分證號碼還是手機號碼的格式是否正確。
#*
#*               (3). 您的工作僅限於負責驗證輸入格式是否正確，不涉及客戶登入的資料內容與後端資料庫內的個資設定是否相同的認證問題。
#*
#*               (4). 當您完成上述工作之後，您只要輸出【身分證號碼有效格式】、【手機號碼有效格式】或【登入錯誤 – 無效格式】其中一個結果於螢幕上。
#*
#*	Homework : No.1
#*
#*      Limitations : (1). 本程式有0輸入錯誤判斷機制，例如輸入是非整數
#*                        的資料物件型態。
#*
#*	Last Modified Date :	2021/03/30
#*
#**********************************************************************


#***********************************************************************
#*                                                                     *
#*                         資料輸入處理區                              *
#*                                                                     *
#***********************************************************************

# ====== 身分證或手機號碼的輸入提示 ====== #

Num = input("請輸入個人的【身分證字號】或【手機號碼】：")

str(Num)                     #強制轉換Num型態為字串，因為輸入手機號碼會被認為是Int型態

# ====== 判斷輸入字元是否為10個字 ====== #

if len(Num) != 10 :                           #利用len()算出字串長度
  print("【登入錯誤 – 無效格式】")              #如不符合以上條件則顯示錯誤，並結束程式
  print()
  input('Press Enter key to exit....')

# ====== 判斷身分證字號及手機號碼及輸出結果====== #

if ord(Num[0]) >= 65 and ord(Num[0]) <= 90 or ord(Num[0]) >= 97 and ord(Num[0]) <=122: #利用轉換ascii code判斷第一個字是否為英文大小寫
    num =''.join([x for x in Num if x.isdigit()])   #抓出Num字串裡的數字
    if len(num) == 9 and num[0] == '1' or num[0] == '2':     #並判斷是否為9個數字以及判斷身分證字號的第一個數字是否為1或是2
        print("【身分證號碼有效格式】")               #如符合以上條件則顯示輸入有效
        print()
        input('Press Enter key to exit....')
    else:                                     #如不符合以上條件則顯示錯誤，並結束程式
        print("【登入錯誤 – 無效格式】") 
        print()
        input('Press Enter key to exit....')    
elif Num[0] =='0':                            #判斷第一個字是否為數字'0'
    if Num.isdigit() == True:                 #判斷字串裡是否全為數字
        print("【手機號碼有效格式】")           #如符合以上條件則顯示輸入有效
        print()
        input('Press Enter key to exit....')
    else:
        print("【登入錯誤 – 無效格式】")        #如不符合以上條件則顯示錯誤，並結束程式
        print()
        input('Press Enter key to exit....')
  
```
