#### P1
**有一個串列scores紀錄James的比賽得分(如下)。請設計一個程式，使用者可輸入一個整數，之後此程式可以列出James有多少場次得分大於或等於上述輸入整數的分數，本程式限定使用串列生成式(List Comprehension)來完成，請勿使用其它方法。
Scores = [ 33, 22, 41, 25, 39, 43, 27, 38, 40]**


```python
scores=[33, 22, 41, 25, 39, 43, 27, 38, 40]    #設定scores串列
num = int(input("請輸入一個整數："))            #輸入整數作為條件
number =[s for s in scores if s>=num]          #利用串列生成式得結果
print('有',len(number),'場大於等於%d分' %(num))  #輸出
```

#### P2
**有一個字典Players，Players中的每一筆資料包含球員名字和身高資料。請設計一個程式，使用者可輸入一個整數，之後此程式可以於螢幕上列出所有身高(公分)為上述輸入整數(含)以上的球員個人資料，本程式限定使用串列生成式(List Comprehension)來完成，請勿使用其它方法。**
<h4 style="text-align:center">Players = { 'James' : 202, 'Curry' : 193, 'Durant': 205, 'Jordan': 199, 'David' : 211, 'Edie'  : 195  }</h4>


```python
Players = {'James':202, 'Curry':193, 'Durant':205, 'Jordan':199, 'David':211, 'Edie':195 }#設定Players字典
num = int(input("請輸入一個整數："))                     #輸入整數作為條件
list1 =[k for k,v in Players.items() if v>=num]         #利用串列生成式將keys抓出
list2 =[v for k,v in Players.items() if v>=num]         #利用串列生成式將values抓出
list3 = {key:value for key, value in zip(list1,list2)}    #利用串列生成式將list1和list2結合
print(list3)      #輸出結合後字典
```

#### P4
**限定使用 yield 的方式來撰寫一個函式生成器(function generator)，其功能為輸入一個正整數 maxIntVal 與一個當作除數的正整數 divisor，而當每一次呼叫您所撰寫的函式生成器 div() 時，可以依序產生一個介於 1 與 maxIntVal (含)之間所有可以被 divisor 整除的正整數。本題請參考hw3_4.exe執行後的結果。**

```python
maxIntVal = int(input("請輸入一個非0的正整數："))        #輸入非0的正整數為被除數
divisor = int(input("請輸入一個非0的正整數當作除數："))   #輸入一個非0的正整數當作除數
MaxIntVal = maxIntVal + 1     #將輸入的被除數+1，因應for迴圈range的定義。ex:for i in range(x,y) for迴圈程式只會代入x到y-1，不會代入y
def div(d,m):                 #處理除法的函式生成器
    x = d                     # x等於除數
    while True:               #使用while和if迴圈來處理整除的整數和方便做break動作
        if x%d == 0:          #如果x可以被整除，則使用yield將x丟出
            yield x    
        x += 1                #並將x+1
        if x == m:            #當跑完最後一圈時，x等於被除數，再break出整個while迴圈
            break
            
print("以下為從1到%d之間，可以被%d整除的所有整數:" %(maxIntVal,divisor))  #輸出行            
for i in div(divisor,MaxIntVal):    #利用for迴圈執行div函式，將丟出的數印出
    print(i)
print()                             #輸出行
print("程式結束，請按Enter鍵離開...") #輸出行 
```
