#### **共用x軸的多軸圖表 – 加圖例(各別顯示)**

```python
x = np.linspace(0, 10, 50)

sinus = np.sin(x)

sinhs = np.sinh(x)

fig, ax = plt.subplots()

ax.plot(x, sinus, "r-o", label = "Sin(x)")

ax.set_xlabel("x", color = "green")

ax.set_ylabel("Sin(x)", color = "red")

ax.legend(loc = "best")

ax2 = ax.twinx()

ax2.plot(x, sinhs, "g--", label = "Sinh(x)")

ax2.set_ylabel("Sinh(x)", color = "blue")

ax2.legend(loc = "best")

plt.show()
```

#### **共用x軸的多軸圖表 – 顯示單一圖例**

```python
x = np.linspace(0, 10, 50)

sinus = np.sin(x)

sinhs = np.sinh(x)

fig, ax = plt.subplots()

lns1 = ax.plot(x, sinus, "r-o", label = "Sin(x)")

ax.set_xlabel("x", color = "green")

ax.set_ylabel("Sin(x)", color = "red")

ax2 = ax.twinx()

lns2 = ax2.plot(x, sinhs, "g--", label = "Sinh(x)")

ax2.set_ylabel("Sinh(x)", color = "blue")

# 自行建立圖例來顯示所有標籤
lns = lns1 + lns2

labs = [l.get_label() for l in lns]

ax.legend(lns, labs, loc = "best")

plt.show()
```

#### **更改圖表的外觀 – 更改圖表的外觀**

```python
x = np.linspace(0, 10, 50)

sinus = np.sin(x)

sinhs = np.sinh(x)

fig, ax = plt.subplots()

ax.plot(x, sinus, "r-o")

ax.set_xlabel("x", color = "green")

ax.set_ylabel("Sin(x)", color = "red")

ax2 = ax.twinx()

ax2.plot(x, sinhs, "g--")

ax2.set_ylabel("Sinh(x)", color = "blue")

# 指定圖表標題文字
ax.set_title("Sin and Sinh Waves", fontsize="large")

# 更改刻度的外觀
for tick in ax.xaxis.get_ticklabels():
    
    tick.set_fontsize("large")
    
    tick.set_fontname("Times New Roman")
    
    tick.set_color("blue")
    
    tick.set_weight("bold")   

plt.show()
```

#### **Sample 3D作圖**

```python
import numpy as np
from mpl_toolkits.mplot3d.axes3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
#from matplotlib.mlab import bivariate_normal

plt.rcParams['axes.unicode_minus'] = False  # 解決座標軸負數的負號顯示問題，沒這行負號無法正常顯示。

def bivariate_normal(X, Y, sigmax=1.0, sigmay=1.0,
                     mux=0.0, muy=0.0, sigmaxy=0.0):
    """
    Bivariate Gaussian distribution for equal shape *X*, *Y*.
    See `bivariate normal
    <http://mathworld.wolfram.com/BivariateNormalDistribution.html>`_
    at mathworld.
    """
    Xmu = X - mux
    
    Ymu = Y - muy

    rho = sigmaxy / (sigmax * sigmay)
    
    z = Xmu**2 / sigmax**2 + Ymu**2 / sigmay**2 - 2 * rho * Xmu * Ymu / (sigmax * sigmay)
    
    denom = 2 * np.pi * sigmax * sigmay * np.sqrt(1 - rho**2)
    
    return np.exp(-z /(2 * (1 - rho**2))) / denom



# 製作2維格子
N = 200

x = np.linspace(-3.0, 3.0, N)

y = np.linspace(-2.0, 2.0, N)

X, Y = np.meshgrid(x, y)


# 以2變量常態分佈製作2維分佈資料
z = 15 * (bivariate_normal(X, Y, 1.0, 1.0, 0.0, 0.0) -
          bivariate_normal(X, Y, 1.5, 0.5, 1, 1))


# 將資料立體繪製
fig = plt.figure(1)  # (1)

ax = fig.add_subplot(111, projection = '3d')  # (2)

surf = ax.plot_surface(X, Y, z, cmap = cm.Oranges)

ax.set_zlim3d(-3.01, 3.01)

plt.colorbar(surf, shrink = 0.5, aspect = 10)

plt.show()
```
