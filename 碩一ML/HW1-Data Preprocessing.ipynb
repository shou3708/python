```python
import numpy as np
import requests
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup 
```

```python
#宣告month, high, low三個list
month,high,low = [],[],[]
url = 'https://www.twse.com.tw/exchangeReport/FMSRFK?response=html&date=20200101&stockNo=2330'

#使用requests get url
Response= requests.get('https://www.twse.com.tw/exchangeReport/FMSRFK?response=html&date=20200101&stockNo=2330')


#使用BeautifulSoup解讀
sp = BeautifulSoup(Response.text,'lxml')
#print(sp.prettify())

#抓出table，並存成變數datas
datas = sp.select("table")[0]

#在datas中選擇tbody tr的標籤
Tr_data=datas.find("tbody").find_all("tr")
#print(Tr_data)

#使用迴圈找出td，讀取出每個月的資料
#td的第一個欄位為年份，不管他
#td的第二個欄位為月份，加入到month list中
#td的第三個欄位為當月最高價，加入到high list中
#td的第三個欄位為當月最低價，加入到low list中
for i in Tr_data:
    Td = i("td")
    month.append(Td[1].string)
    high.append(Td[2].string)
    low.append(Td[3].string)
    
month =list(map(int, month))
high =list(map(float, high))
low =list(map(float, low))

print(month)
#print(high)
#print(low)

#將圖表顯示出來
plt.plot(high, label = "Highest")
plt.plot(low, label = "Lowest")

plt.title("2330")
plt.xlabel("Month")       
plt.ylabel("Money")   
plt.legend(loc = 2)

plt.show()
```
